```python
from datasets import load_dataset
datasets = load_dataset("./wisesight_sentiment", trust_remote_code=True)
```

```python
train_df = datasets['train'].to_pandas()
val_df = datasets['validation'].to_pandas()
test_df = datasets['test'].to_pandas()
```

```python
# Using a mapping dictionary
mapping = {
    0: "positive",
    1: "neutral",
    2: "negative"
}
```

```python
train_df["label"] = train_df["category"].apply(lambda x: mapping.get(x, "q"))
train_df = train_df[train_df['label'] != 'q'].reset_index(drop=True)

val_df["label"] = val_df["category"].apply(lambda x: mapping.get(x, "q"))
val_df = val_df[val_df['label'] != 'q'].reset_index(drop=True)

test_df["label"] = test_df["category"].apply(lambda x: mapping.get(x, "q"))
test_df = test_df[test_df['label'] != 'q'].reset_index(drop=True)
```

```python
train_sample = train_df.drop_duplicates().sample(n=10000, random_state=42)
val_sample = val_df.drop_duplicates().sample(n=1000, random_state=42)
test_sample = test_df.drop_duplicates().sample(n=2000, random_state=42)
```

```python
SYSTEM_PROMPT = """You are an AI assistant specialized in social sentiment analysis.
Your task is to classify the sentiment expressed in a given piece of text from social media into one of the following categories:

positive: The text expresses favorable, supportive, or happy emotions.
negative: The text expresses unfavorable, critical, or unhappy emotions.
neutral: The text expresses neither positive nor negative sentiment, or it is objective, factual, or ambiguous.
"""
```

```python
train_data_openai_format = []
val_data_openai_format = []
test_data_openai_format = []
```

```python
for row in train_sample.iterrows():
    train_data_openai_format.append(
        {
        "messages": [
            {
                "role": "system",
                "content": SYSTEM_PROMPT
            },
            {
                "role": "user",
                "content": row[1]['texts']
            },
            {
                "role": "assistant",
                "content":  row[1]['label']
            }
        ]}
    )
```

```python
for row in val_sample.iterrows():
    val_data_openai_format.append(
        {
        "messages": [
            {
                "role": "system",
                "content": SYSTEM_PROMPT
            },
            {
                "role": "user",
                "content": row[1]['texts']
            },
            {
                "role": "assistant",
                "content":  row[1]['label']
            }
        ]}
    )
```

```python
for row in test_sample.iterrows():
    test_data_openai_format.append(
        {
        "messages": [
            {
                "role": "system",
                "content": SYSTEM_PROMPT
            },
            {
                "role": "user",
                "content": row[1]['texts']
            },
            {
                "role": "assistant",
                "content":  row[1]['label']
            }
        ]}
    )
```
